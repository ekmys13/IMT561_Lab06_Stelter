```python
# plot test histogram
fig = px.histogram(df, x='response_time_days', nbins=30,
                   title="Distribution of Response Time in Days",
                   labels={'response_time_days':'Response Time (Days)'
                   })
fig
# Interpretation sentence (for layouts): The distribution of response time in days is heavily right-skewed. The average response time for each complaint is 6.17 days, and the median response time is 5 days.
```

```python
# plot test borough bar chart

# aggregating by dates
agg = (
        df.groupby("borough", as_index=False)["response_time_days"]
        .median()
        .rename(columns={"response_time_days": "median_response_days"})
        .sort_values("median_response_days", ascending=False) # ascending_false sets it so we're sorting the values from largest to smallest
    )

fig = px.bar(agg, x="borough", y="median_response_days", title="Response Time by Borough",)

fig
```

```python
# plot test chart - count by boroughs

# aggregating by borough
borough_counts = df.groupby("borough").size().sort_values(ascending=False).to_frame()
borough_counts = borough_counts.rename(columns={0:"Number of Complaints"})
borough_counts

fig = px.bar(borough_counts, y="Number of Complaints", title="Number of Complaints by Borough")

fig
```

```python
# plot test complaint type bar chart

# aggregating by response_time_days
agg = (
        df.groupby("complaint_type", as_index=False)["response_time_days"]
        .median()
        .rename(columns={"response_time_days": "median_response_days"})
        .sort_values("median_response_days", ascending=False) # ascending_false sets it so we're sorting the values from largest to smallest
    )

fig = px.bar(agg, x="complaint_type", y="median_response_days", title="Response Time by Complaint Type",)

fig
```
